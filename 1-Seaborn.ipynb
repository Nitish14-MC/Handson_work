# SEABORN::

* Seaborn is a data visualisation library



* It provides plots more attractive



* It is built on top of matplotlib



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
mtcars=pd.read_csv("https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv")
```

```python
mtcars
```

```python
mtcars.head()
```

```python
mtcars.info()
```

```python
mtcars.shape
```

```python
# BAR PLOT::
res=sns.barplot(mtcars['cyl'],mtcars['mpg'])
plt.show()
```

```python
# OTHER WAY OF CREATING THE ABOVE

res=sns.barplot(x='cyl',y='mpg',data=mtcars)
plt.show()
```

# Here I have not mentioned any colors ...by default it's taking different colors for every bar...This is why it is called advanced visualiastion tool

```python
# Change the color
res=sns.barplot(x='cyl',y='mpg',data=mtcars,color='yellow')
plt.show()
```

# Seaborn palettes::
Seaborn makes it easy to use colors that are well-suited to the characteristics of your data and your visualization goals


https://seaborn.pydata.org/tutorial/color_palettes.html

```python
# Using palette::::


res=sns.barplot(x='cyl',y='mpg',data=mtcars,palette="Set2")
plt.show()
```

```python
res=sns.barplot(x='cyl',y='gear',data=mtcars,palette="Set2")
plt.show()
```

```python
res=sns.barplot(x='cyl',y='gear',data=mtcars,palette="rocket")
plt.show()
```

## Countplot::

```python
sns.countplot(x='carb',data=mtcars,palette="Set1")
```

```python
sns.countplot(x='carb',data=mtcars,palette="Set2")
```

### How to create a horizontal plot

```python
sns.countplot(y='carb',data=mtcars,palette="Set2")
```

### Using hue parameter ::

```python
sns.countplot(x='gear',hue='cyl',data=mtcars,palette="Set1")
```

## Distribution plot---->distplot

```python
sns.distplot(mtcars.mpg,bins=10,color='g')
```

# Creating heatmap

```python
sns.heatmap(mtcars.corr(),cbar=True,linewidth=0.5)
```

# IRIS DATASET:: 
The Iris Dataset contains four features (length and width of sepals and petals) of 50 samples of three species of Iris (Iris setosa, Iris virginica and Iris versicolor).

```python
iris=sns.load_dataset('iris')
```
