# How to increase the size of datapoint::

```python
x=[40,50,67,78,91,52]
y=[78,89,45,65,42,12]
plt.plot(x,y,marker='X',ms="15")# ms=marker size
plt.title("SIMPLE LINE PLOT")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

# COLORING MARKER EDGES::

```python
x=[40,50,67,78,91,52]
y=[78,89,45,65,42,12]
plt.plot(x,y,marker='s',mec="r")# s-for square , mec---colors the edges to red
plt.title("SIMPLE LINE PLOT")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

# Coloring the FACE EDGES::

```python
x=[40,50,67,78,91,52]
y=[78,89,45,65,42,12]
plt.plot(x,y,marker='s',mfc="r",ms=15)# s-for square , mfc---colors the face to red , ms=marker size 
plt.title("SIMPLE LINE PLOT")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
x=[40,50,67,78,91,52]
y=[78,89,45,65,42,12]
plt.plot(x,y,marker='*',mfc="r",ms=15)# s-for square , mfc---colors the face to red , ms=marker size 
plt.title("SIMPLE LINE PLOT")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

# GRID FUNCTION WITH PROPERTIES::

1.Grid lines::


2.Grids can be applied with respect to axis::

ONLY ON X-AXIS


ONLY ON Y-AXIS


GRIDS ON BOTH AXIS::


GRID LINES HAS PROPERTIES:


SYNTAX FOR GRID:



```python
x=np.arange(0,20,4)
y=x**2
plt.plot(x,y)
plt.grid() # By default grids are applied on both axis
plt.show()
```

```python
x=np.arange(0,20,4)
y=x**2
plt.plot(x,y)
plt.grid(axis="x") # grids  applied on X- axis
plt.show()
```

```python
x=np.arange(0,20,4)
y=x**2
plt.plot(x,y)
plt.grid(axis="y") # grids  applied on Y- axis
plt.show()
```

```python
x=np.arange(0,20,4)
y=x**2
plt.plot(x,y)
plt.grid(axis="both") # grids  applied on both axis
plt.show()
```

#### Grid line properties::

```python
x=np.arange(0,20,4)
y=x**2
plt.plot(x,y)
plt.grid(axis="both",ls="dotted",lw=2,c="green") # grids  applied on Y- axis
plt.show()
```

4.Syntax for column subplot::

```python
x=np.arange(0,10,4)
y=x**2
y1=x**3

# First plot
plt.subplot(1,2,1)
plt.plot(x,y)

# Second plot
plt.subplot(1,2,2)
plt.plot(x,y1)

plt.show()
```

### We will get an error

```python
x=np.arange(0,10,4)
y=x**2
y1=x**3

plt.subplot(1,2,1) #(1-Row,2-Columns,1-Index)
# Here we have 1-row and 2 columns....so totally 1x2=2 subplots possible
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,y1)

plt.subplot(1,2,3)# Here maximum only 2-subplots are possible.....But you ar trying to give 3-index....which will raise error
plt.plot(x,y1)

plt.show()
```

## Lets plot 2-rows and 2-columns------IN TOTAL 4-SUBPLOTS

```python
x=np.arange(0,10,4)
y=x**2
y1=x**3

#first plot
plt.subplot(2,2,1) #(1-Row,2-Columns,1-Index)
# Here we have 1-row and 2 columns....so totally 1x2=2 subplots possible
plt.plot(x,y)

# Second plot
plt.subplot(2,2,2)
plt.plot(x,y1)

# Third plot
plt.subplot(2,2,3)# Here maximum only 2-subplots are possible.....But you ar trying to give 3-index....which will raise error
plt.plot(x,y1)

#Fourth plot
plt.subplot(2,2,4)
plt.plot(x,y1)

plt.show()
```

# Using TITLES for subplots ..

```python
x=np.arange(0,10,4)
y=x**2
y1=x**3

# MAIN TITLE: 
plt.suptitle("MULTIPLE PLOTS")

#first plot
plt.subplot(2,2,1) #(1-Row,2-Columns,1-Index)
# Here we have 1-row and 2 columns....so totally 1x2=2 subplots possible
plt.plot(x,y)
plt.title("PLOT-1")

# Second plot
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("PLOT-2")

# Third plot
plt.subplot(2,2,3)# Here maximum only 2-subplots are possible.....But you ar trying to give 3-index....which will raise error
plt.plot(x,y1)
plt.title("PLOT-3")

#Fourth plot
plt.subplot(2,2,4)
plt.plot(x,y1)
plt.title("PLOT-4")

plt.show()
```

# PIE-CHART::
A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

```python
from matplotlib import pyplot as plt
```

```python
student_performance=["Excellent","Good","Average","Poor"]
student_values=[20,15,10,5]
plt.pie(student_values,labels=student_performance)
plt.show()
```

```python
student_performance=["Excellent","Good","Average","Poor"]
student_values=[20,15,10,5]
plt.pie(student_values,labels=student_performance,startangle=180)
plt.show()
```

```python
# USing explode parameter
student_performance=["Excellent","Good","Average","Poor"]
student_values=[20,15,10,5]
plt.pie(student_values,labels=student_performance,startangle=360,explode=[0.2,0,0,0])
plt.show()
```

```python
student_performance=["Excellent","Good","Average","Poor"]
student_values=[20,15,10,5]
plt.pie(student_values,labels=student_performance,startangle=360,explode=[0.2,0.1,0.3,0.5])
plt.show()
```

```python
# using colors parameters
student_performance=["Excellent","Good","Average","Poor"]
student_values=[20,15,10,5]
plt.pie(student_values,labels=student_performance,startangle=360,explode=[0.2,0.1,0.3,0.5],colors=['red','yellow','black','blue'])
plt.show()
```

```python
# using percentage parameters
student_performance=["Excellent","Good","Average","Poor"]
student_values=[20,15,10,5]
plt.pie(student_values,labels=student_performance,startangle=360,explode=[0.2,0.1,0.3,0.5],colors=['red','yellow','brown','blue'],autopct="%2.1f%%")
plt.show()
```

```python
# Using figure size & Legend


student_performance=["Excellent","Good","Average","Poor"]
student_values=[20,15,10,5]
plt.figure(figsize=(8,10))
plt.pie(student_values,labels=student_performance,startangle=360,explode=[0.2,0.1,0.3,0.5],colors=['red','yellow','brown','blue'],autopct="%2.1f%%")
plt.legend(title="Performance")
plt.show()
```
