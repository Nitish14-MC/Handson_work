# Working on JSON Data:

* JSON stands for Javascript Object Notation. JSON is a text-based data format that is used to store and transfer data. For example, // JSON syntax { "name": "John", "age": 22, "gender": "male", } In JSON, the data are in key/value pairs separated by a comma , . JSON was derived from JavaScript.

```python
s="""
{
"name":"Nitish",
"Age":"20",
"College":"Srm University",
"Specialisation":["Ai","Ml","Dl","Nlp","Bd"]
}
"""
```

```python
import json
```

```python
s1=json.loads(s)
```

```python
s1
```

# How to convert json data into tabular format

```python
pd.DataFrame(s1)
```

```python
# I need only the "Specialisation column"
pd.DataFrame(s1,columns=['Specialisation'])
```
