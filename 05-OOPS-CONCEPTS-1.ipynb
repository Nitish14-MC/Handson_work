## USE SELF::

* The self variable is used to represent the instance of the class which is often used in object-oriented programming.

* It works as a reference to the object or returns current object location

* Python uses the self parameter to refer to instance attributes and methods of the class. 

```python
class Bottle:
    #Properties/state
    id=1
    color='red'
    capacity=1
    height=30

    # functions/behaviour
    def wash(self): # Here instead of self we can even use some other name....but it will even be considered as memory address
        print(self)
        print('Washing')
    
    def setcap(self):
        print(self)
        print('Setting cap')
    
    def fillwater(self):
        print(self)
        print('fill water')
```

```python
bottle1=Bottle()
print(bottle1)
```

```python
bottle1.wash()
```

# Class variables and instance variables::

* Class variables:: Class variables are shared across all objects

* Instance variables:: while instance variables are for data unique to each instance.

## Note:: Instance variable overrides the Class variables having same name which can accidentally introduce bugs or surprising behaviour in our code.

```python
# Example for class level variables
class Bottle:
    #Properties/state
    # THESE ARE CALLED CLASS LEVEL VARIABLES
    id=1
    color='red'
    capacity=1
    height=30
 
    # Constructor
    def __init__(self,color,capacity):
        self.color=color
        self.capacity=capacity
        print('constructor')

    # functions/behaviour
    def wash(self):
        #print(self)  Commenting this part
        print('Washing')
    
    def setcap(self):
        #print(self) Commenting this part
        print('Setting cap')
    
    def fillwater(self):
        #print(self) Commenting this part
        print('fill water')
```

```python
# Example for instance level variables
class Bottle:
    #Properties/state
    id=1
    color='red'
    capacity=1
    height=30
 
    # Constructor
    def __init__(self,color,capacity):
        self.color=color # THESE ARE CALLED CLASS LEVEL VARIABLES
        self.capacity=capacity # THESE ARE CALLED CLASS LEVEL VARIABLES
        print('constructor')

    # functions/behaviour
    def wash(self):
        #print(self)  Commenting this part
        print('Washing')
    
    def setcap(self):
        #print(self) Commenting this part
        print('Setting cap')
    
    def fillwater(self):
        #print(self) Commenting this part
        print('fill water')
```

# Destructor:

* Destructors are called when object gets destroyed

```python
# Example::
class Bottle:
    #Class variables::
    companyName="Sirpa"
    
 
    # Constructor
    def __init__(self,color,capacity):
        self.color=color # THESE ARE CALLED CLASS LEVEL VARIABLES
        self.capacity=capacity # THESE ARE CALLED CLASS LEVEL VARIABLES
        print('constructor')
        
        
    def __del__(self):
        print("destructor")

    # functions/behaviour
    def wash(self):
        #print(self)  Commenting this part
        print('Washing')
    
    def setcap(self):
        #print(self) Commenting this part
        print('Setting cap')
    
    def fillwater(self):
        #print(self) Commenting this part
        print('fill water')
```

```python
bottle1=Bottle('Red',1)
bottle2=Bottle("Grey",9)
bottle1.wash()
print(bottle1.color)
```
