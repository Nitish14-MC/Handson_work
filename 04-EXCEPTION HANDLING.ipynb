# What are errors ??::

* Errors are the problems in a program due to which the program will stop the execution.

* Types of errors: a)syntax errors and b) Logical errors

* Logical errors are called exceptions

## Difference between syntax errors and logical errors

### Syntax errors:

* program stops execution

### Logical errors(Exceptions):

* Unexpected behaviour

# Example for syntax error:

```python
class Join:
    def __init__(self) # Here we don't have ":",it will rise an syntax error
        print('Join with us')
        
    def welcome(self):
        print('Welcome to our team')
        
j1=Join()
j1.welcome()
```

# Note:: In syntax error the program will not excecute completly..it will stop at the place wherever there is an error.

## Exception::

* No problem in the syntax(syntactically correct) but there is a problem in the code

### Identify the exception in the below program

```python
class Join:
    def __init__(self): # Here we don't have ":",it will rise an syntax error
        print('Join with us')
        
    def welcome(self):
        print('Welcome to our team')
        
    def __del__(self):
        print("destructor")
        
    def members(self):
        members=['Tom','Ram','Sam']
        print(members[3]) # Here it is out of range
        
j1=Join()
j1.welcome()
j1.members()
```

# How to handle above mentioned exceptions::

* We use Try and Except

```python
# Using try and except
class Join:
    def __init__(self): # Here we don't have ":",it will rise an syntax error
        print('Join with us')
        
    def welcome(self):
        print('Welcome to our team')
        
    
        
    def members(self):
        try:
            members=['Tom','Ram','Sam']
            print(members[4]) # Here it is out of range
        except:
            print("Exception occured")
        
    
        
j1=Join()
j1.welcome()
j1.members()
```

## How to know what is the exact info about exception

* Make use of sys library

```python
import sys
class Join:
    def __init__(self): # Here we don't have ":",it will rise an syntax error
        print('Join with us')
        
    def welcome(self):
        print('Welcome to our team')
        
    
        
    def members(self):
        try:
            members=['Tom','Ram','Sam']
            print(members[4]) # Here it is out of range
        except:
            errorInfo=sys.exc_info()
            print(errorInfo[0],errorInfo[1])
        
    
        
j1=Join()
j1.welcome()
j1.members()
```
