# Name feature


* The name feature in the pandas dataframe gives the column name by default it would be indexed

```python
# We can over-ride as per our wish
# here pass a list of name variables
pd.read_csv('deliveries.csv',sep='#',header=None,names=['A','B','C','D','E','F','G','H','I','J','K'])
```

# na_value:

* This is another important feature

```python
df2
```
