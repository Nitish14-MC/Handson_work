# Program-3: Write a program to check whether a string is palindrome?

* What is palindrome:
* a word, phrase, or sequence that reads the same backwards as forwards

```python
name=input("Enter your name::")

print("Entered name is :: ",name.upper())
reverse='' # creating an empty string here
for i in name:
    reverse=i+reverse
print("The reversed value of {} is {}".format(name,reverse))

if(name==reverse):
    print("It is a palindrome")
else:
    print("Not a palindrome")
```

# Program-04: Write a program to find factorial of a number 

* 5 factorial is 5 x 4 x 3 x 2 x 1 = 120

```python
#logic:

# Here core concept is multiplication ie, 5! =120 ie, 5x4x3x2x1=1x2x3x4x5
# Consider n=4 ie, here n+1=5 ,so loop iterates 4 times
# mul =1
# when i=1=====> mul x i =1 x 1 =1---> ie,mul= 1
# when i=2=====>mul x i = 1 x 2=2---->ie,mul=2
#when i=3=====>mul x i=2x3==>6--->ie,mul=6
#when i=4===>mul x i =6 x 4 = 24--->ie mul=24
```

```python
n=int(input("Enter the number::"))
if (n<=0):
    print("Invalid number")
elif(n==0):
    print("Factorial of 0 is 1")
else:
    mul=1
    for i in range(1,n+1):
        mul=mul*i
    print("Factorial of {} is {}".format(n,mul))
```
